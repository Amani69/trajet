```
def get_date(data):
  data=(data.withColumn("date", to_date(from_unixtime(col('dateentry')/1000)))
                     .withColumn("month",month("date"))
                     .withColumn("week",weekofyear("date"))
                     .withColumn("year",year("date"))
                     )
  return data
data_date=get_date(data)
```

```
data_date.printSchema()
```

```
#groupby trajet
#garder seulement le point de depart de chaque trajet (pour ajouter le code postal)
#garder la date debut de chaque trajet
#calcule la duree de trajet  partir la colonne dateentry
#pour chaque trajet a un seul device 

def groupby_trajet(data):
  DF=data.groupby("rideid").agg(
    first("deviceid").alias("deviceid"),
    first("longitude").alias("long_start"),
    first("latitude").alias("lat_start"),
    sum("distance").alias("distance_trajet"),
    min("date").alias('date_trajet'),
    ((max("dateentry")-min("dateentry"))/1000 ).alias("duree_trajet"))
  return DF
data_trajet=groupby_trajet(data_date)
```

```
data_2018_10=data_trajet.filter((year("date_trajet")==2018) & (month("date_trajet").isin([10])))
```

```
data_2018_10.display()
```

```
df_2018_device=data_2018_10.groupby("date_trajet","deviceid").agg(
                                            count_distinct("rideid").alias('nombre_trjet'),
                                              sum("distance_trajet").alias('nombre_km'),
                                              sum("duree_trajet").alias('duree_trajet'))
```

```
df_filter=df_2018_device.filter((col("date_trajet")>="2018-10-08") & (col("date_trajet")<="2018-10-14"))
df_filter.display()
```
