```
def get_year(df,value_):
  df = df.withColumn('year',F.year("date"))
  if type(value_)!= list:
    val=[]
    val.append(value_)
  else : 
    val = value_

  df=df.filter(F.col('year').isin(val))
  return df 
```

```
def get_month(df):
  df = df.withColumn('month',F.month("date"))
  return df 
```

```
def get_week_of_year(df):
  df = df.withColumn("Week_of_the_year",F.weekofyear( "date"))
  return df 
```

```
def get_day(df,value_):
  df=df.withColumn('day', F.dayofmonth("date"))
  if type(value_)!= list:
    val=[]
    val.append(value_)
  else : 
    val = value_
  df=df.filter(F.col('day').isin(val))
  return df 
```

```
def get_dayofweek(df):
  df=df.withColumn("dayofweek", F.date_format("date", "EEEE"))
  return df
```

```
def get_quarter(df):
  df=df.withColumn("quarter", F.quarter("date"))
  return df
```

```
#un trajet a t effectu plus de 2 fois par semaine sur une fentre de 3 mois 
```

```
def frequency(df,partition,group):
  windows = Window.partitionBy(partition).orderBy(F.col("date").asc())
  df = (
  df.withColumn('rideid', F.first('rideid').over(windows))
    .withColumn('deviceid', F.first('deviceid').over(windows))
      )
  df2=df.groupBy("type_ride_id",group).agg(F.count("deviceid").alias('frequence'),F.avg('vitesse_moy').alias('vitess_moyenne'),F.avg("distance_trajet").alias("distance_moy"))
  return  df2
```
